## Question-1

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


f1 = pd.read_csv('logisticX.csv', names = ['x1','x2'])
f2 = pd.read_csv('logisticY.csv', names = ['y'])

x1 = f1['x1']
x2 = f1['x2']
y = f2['y']

data = np.zeros((y.shape[0], 3))

data[ : , 0] = x1
data[ : , 1] = x2
data[ : , 2] = y

np.random.shuffle(data)


def hypothesis(data, theta):
    
    h = data[1] * theta[2] + data[0] * theta[1] + theta[0]
    return sigmoid(h)


def sigmoid(z):
    return 1/(1 + np.exp(-1*z)) 


def error(data, theta):
    
    err = 0
    
    for i in range(data.shape[0]):
        
        h = hypothesis(data[i][ :2], theta)
        err += data[i][2] * np.log2(h) + (1 - data[i][2]) * np.log2(1 - h)
        
    return -err


def cal_grad(data, theta):
    
    grad = np.zeros(data.shape[1])
    
    indices = np.arange(data.shape[0])
    np.random.shuffle(indices)
    indices = indices[:10]

    for i in indices:
            
        h = hypothesis(data[i][ :2], theta)
        
        grad[1:] += (data[i][2] - h) * data[i][ :2]
        grad[0] += (data[i][2] - h)
        
    grad/=data.shape[0]
    
    return grad
    


def gradient_des(data, learning_rate):
    
    error_list = []
    theta = np.zeros(data.shape[1])
    
    for i in range(300):
        
        total_error = error(data, theta)
        error_list.append(total_error)
        
        grad = cal_grad(data, theta)
        
        theta = theta + learning_rate * grad
        
    return theta, error_list


theta, err = gradient_des(data, learning_rate = 0.5)

plt.style.use('seaborn')
plt.plot(err)
plt.show()
```

```python
err = np.array(err)

size = err.shape[0]
r = int(size * 80 / 100)

avg_err = np.sum(err[r:])/(size * 20 / 100)
print("Average error = %d" %avg_err)

print("\ncofficients :")
print("theta[0] = %.4f (intercept)" %theta[0])
print("theta[1] = %.4f " %theta[1])
print("theta[2] = %.4f " %theta[2])
```
