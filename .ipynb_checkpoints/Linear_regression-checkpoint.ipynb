## Linear Regrassion using function (linregress())

```python
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import make_regression

x,y = make_regression(n_features=1,n_samples=30,bias=4,noise=8)

x = x.reshape((-1,))
y = y.reshape((-1,))

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

plt.plot(x, y,'o', label = 'original data')
plt.plot(x, intercept + slope*x, 'r', label = 'fitted line')

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.show()
```

## Linear Regression using gradient descent

```python
import matplotlib.pyplot as plt
import numpy as np


def error(x, y, theta):
    
    total_error = 0
    
    for i in range (len(x)):
        total_error += (theta[1] * x[i] + theta[0] - y[i])**2

    return total_error
        
        
def cal_gradient(x, y, theta):
    
    grad = np.zeros((2,))
    
    for i in range(len(x)):
        
        grad[0] += (theta[1] * x[i] + theta[0] - y[i])
        grad[1] += (theta[1] * x[i] + theta[0] - y[i]) * x[i]
                   
    return grad


def gradientDescent(x, y):
    
    iterations = 500
    error_list = []
    theta = np.zeros((2,))
    
    while iterations > 0:
        
        total_error = error(x, y, theta)
        error_list.append(total_error)
        grad = cal_gradient(x, y, theta)
        
        theta = theta - 0.001 * grad
        theta = theta - 0.001 * grad
        
        iterations -=1
    
    return theta, error_list


def drawLine(theta):
    x_pts = np.linspace(-2,5,10)
    y_pts = theta[1] * x_pts + theta[0]

    plt.scatter(x,y)
    plt.plot(x_pts,y_pts,color='black')
    plt.show()
  
x, y = make_regression(n_features=1,n_samples=30,bias=4,noise=80)

x = x.reshape((-1,))
y = y.reshape((-1,))

theta, error_list = gradientDescent(x, y)

plt.plot(error_list)
plt.show()

drawLine(theta)

print(error_list[-1])
```

```python
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import make_regression
import numpy as np
from mpl_toolkits.mplot3d import Axes3D


def hypothesis(x, theta):
    return theta[0] + x * theta[1]


def error(x, y, theta):
    
    total_error = 0
    
    for i in range(x.shape[0]):
        h = hypothesis(x[i],theta)
        total_error += (h - y[i]) ** 2
    
    return total_error


def batch_gradient (x, y, theta):
    
    grad = np.zeros((2,))
    
    for i in range(x.shape[0]):
        h = hypothesis(x[i], theta)
        grad[0] += (h - y[i])
        grad[1] += (h - y[i]) * x[i]
                   
    return grad


def run_gradient (x, y, iterations, gamma):        
    theta = np.array([-2.0,0])
    
    error_list = []
    theta_list = []
    
    while (iterations > 0):
        grad = batch_gradient(x, y, theta)
        theta = theta - gamma * grad
        e = error(x, y, theta)
        
        error_list.append(e)
        theta_list.append((theta[0], theta[1]))
        iterations -=1
    
    return error_list, theta, theta_list


def drawLine(theta):
    x_pts = np.linspace(-2,5,10)
    y_pts = hypothesis(x_pts,theta)

    plt.scatter(x,y)
    plt.plot(x_pts,y_pts,color='black')
    
    
x, y = make_regression(n_features = 1, n_samples = 30, bias = 4, noise = 8)

iterations = 100
gamma = 0.001
e, theta, theta_list = run_gradient(x, y, iterations, gamma)

plt.plot(e)
plt.show()

drawLine(theta)
plt.show()

print("Error is %f"%e[-1])


t0 = np.arange(-2,3,0.01)
t1 = np.arange(-2,3,0.01)

t0,t1 = np.meshgrid(t0,t1)

J = np.zeros(t0.shape)

m = T0.shape[0]
n = T0.shape[1]


for i in range(m):
    for j in range(n):
        J[i,j] = np.sum((y - t1[i,j]*x - t0[i,j])**2)
        
theta_list = np.array(theta_list)        

fig = plt.figure()
axes = fig.gca(projection = "3d")
#axes.scatter(theta_list[:,0],theta_list[:,1],e,c='k')
axes.plot_surface(t0, t1, J, cmap = "rainbow")

plt.show()

fig = plt.figure()
axes = fig.gca(projection = "3d")
axes.scatter(theta_list[:,0], theta_list[:,1],e)

axes.contour(t0, t1, J, cmap = "rainbow")

plt.show()

plt.contour(t0, t1, J)
plt.scatter(theta_list[:,0], theta_list[:,1],c = 'k')

plt.show()
```
