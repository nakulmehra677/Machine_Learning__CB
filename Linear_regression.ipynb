## Linear Regrassion using function (linregress())

```python
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import make_regression

x,y = make_regression(n_features=1,n_samples=30,bias=4,noise=8)

x = x.reshape((-1,))
y = y.reshape((-1,))

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

plt.plot(x, y,'o', label = 'original data')
plt.plot(x, intercept + slope*x, 'r', label = 'fitted line')

plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.legend()
plt.show()
```

## Find local minima

```python
import matplotlib.pyplot as plt

x = []
y = []
f = lambda x: x**2 - 2*x+2
    
for i in range(-20,20,1):
    x.append(i)
    y.append(f(i))

plt.plot(x,y)
plt.title("Real function")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

cur_x = 6 # The algorithm starts at x=6
gamma = 0.01 # step size multiplier
precision = 0.00001
previous_step_size = 1 
max_iters = 10000 # maximum number of iterations
iters = 0 #iteration counter

plt.title("Gradient starting from %d"%cur_x)

df = lambda x: 2*x - 2

x1 = []
y1 = []

while (previous_step_size > precision) & (iters < max_iters):
    prev_x = cur_x
    cur_x -= gamma * df(prev_x)
    previous_step_size = abs(cur_x - prev_x)
    iters+=1
    x1.append(cur_x)
    y1.append(f(cur_x))
   
plt.plot(x1,y1)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()

print("The function is minimum at %f"%cur_x)
    
```

## Linear Regression using gradient descent

```python
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import make_regression
import numpy as np


def hypothesis(x, theta):
    return theta[0] + x * theta[1]


def error(x, y, theta):
    
    total_error = 0
    
    for i in range(len(x)):
        h = hypothesis(x[i],theta)
        total_error += (h - y[i]) ** 2
    
    return total_error


def batch_gradient (x, y, theta):
    
    grad = np.zeros((2,))
    
    for i in range(len(x)):
        h = hypothesis(x[i], theta)
        grad[0] += (h - y[i])
        grad[1] += (h - y[i]) * x[i]
                   
    return grad


def run_gradient (x, y, iterations, gamma):        
    theta = np.zeros((2,))
    
    error_list = []
    
    while (iterations):
        grad = batch_gradient(x, y, theta)
        theta = theta - gamma * grad
        e = error(x, y, theta)
        
        error_list.append(e)
        iterations -=1
    
    return error_list, theta


def drawLine(theta):
    x_pts = np.linspace(-2,5,10)
    y_pts = hypothesis(x_pts,theta)

    plt.scatter(x,y)
    plt.plot(x_pts,y_pts,color='black')
    
    
x, y = make_regression(n_features = 1, n_samples = 30, bias = 4, noise = 8)

iterations = 500
gamma = 0.001
e, theta = run_gradient(x, y, iterations, gamma)

plt.plot(e)
plt.show()

drawLine(theta)
plt.show()

print("Error is %f"%e[-1])
```

```python
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import make_regression
import numpy as np
from mpl_toolkits.mplot3d import Axes3D


def hypothesis(x, theta):
    return theta[0] + x * theta[1]


def error(x, y, theta):
    
    total_error = 0
    
    for i in range(x.shape[0]):
        h = hypothesis(x[i],theta)
        total_error += (h - y[i]) ** 2
    
    return total_error


def batch_gradient (x, y, theta):
    
    grad = np.zeros((2,))
    
    for i in range(x.shape[0]):
        h = hypothesis(x[i], theta)
        grad[0] += (h - y[i])
        grad[1] += (h - y[i]) * x[i]
                   
    return grad


def run_gradient (x, y, iterations, gamma):        
    theta = np.array([-2.0,0])
    
    error_list = []
    theta_list = []
    
    while (iterations > 0):
        grad = batch_gradient(x, y, theta)
        theta = theta - gamma * grad
        e = error(x, y, theta)
        
        error_list.append(e)
        theta_list.append((theta[0], theta[1]))
        iterations -=1
    
    return error_list, theta, theta_list


def drawLine(theta):
    x_pts = np.linspace(-2,5,10)
    y_pts = hypothesis(x_pts,theta)

    plt.scatter(x,y)
    plt.plot(x_pts,y_pts,color='black')
    
    
x, y = make_regression(n_features = 1, n_samples = 300, bias = 4, noise = 8)

iterations = 500
gamma = 0.001
e, theta, theta_list = run_gradient(x, y, iterations, gamma)

plt.plot(e)
plt.show()

drawLine(theta)
plt.show()

print("Error is %f"%e[-1])


t0 = np.arange(theta[0])

t1 = np.arange(theta[1])

t0, t1 = np.meshgrid(t0,t1)
J = np.zeros(t0.shape)

m = t0.shape[0]
n = t0.shape[1]

theta_list = np.array(theta_list)

for i in range(m):
    for j in range(n):
        J[i,j] = np.sum((y - t1[i,j]*x + t0[i,j])**2)
        

fig = plt.figure()
axes = fig.gca(projection = "3d")
axes.scatter(theta_list[:,0], theta_list[:,1],e)
axes.plot_surface(t0, t1, J, cmap = "rainbow")

plt.show()

fig = plt.figure()
axes = fig.gca(projection = "3d")
axes.scatter(theta_list[:,0], theta_list[:,1],e)

axes.contour(t0, t1, J, cmap = "rainbow")

plt.show()

plt.contour(t0, t1, J)
plt.scatter(theta_list[:,0], theta_list[:,1],c = 'k')

plt.show()
```
